```python
%matplotlib inline
import tensorflow as tf
import numpy
import matplotlib.pyplot as plt


batch_size = 10
learning_rate = 0.1
move_avg_decay = 0.99
reg_lambda = 0.001

global_step = tf.Variable(0, trainable=False)

regularization = tf.contrib.layers.l2_regularizer(reg_lambda)

x = tf.placeholder(tf.float32, shape=(None,1))
y_ = tf.placeholder(tf.float32, shape=(None,1))

var_avg = tf.train.ExponentialMovingAverage(move_avg_decay, global_step)
var_avg_op = var_avg.apply(tf.trainable_variables())

def layer(in_, out_, x, avg_class, regular, active):
    w = tf.Variable(tf.random_normal([in_, out_], stddev=0.1))
    b = tf.Variable(tf.random_normal([out_], stddev=0.1))
    tf.add_to_collection('losses', regular(w))
    
    if avg_class != None:
        aw = avg_class.average(w)
        ab = avg_class.average(b)
        print("aw:ab", aw, ab)
        wxb = tf.matmul(x, aw)+ab
        return active(wxb) 
    
    wxb = tf.matmul(x, w) + b
    return active(wxb)
    
#endf layer
y = layer(1, 1, x, var_avg, regularization, tf.nn.sigmoid)



print("ok")
```
